```python
%%sql

select table_name from
information_schema.tables where
table_schema = 'public';
```

```python
%%sql

select now();

# Select Extract(month from now());

select date_trunc('year', now());
```

```python
%%sql

select  column_name from
information_schema.columns where
table_name = 'sales';
```

```python
%%sql

select
    date_trunc('month', orderdate)::date as order_month,
    sum(quantity * netprice * exchangerate) as total_sales_monthly,
    count(Distinct customerkey) as customer_count
from sales
group by
    order_month
order by
    order_month
limit 10;
```
